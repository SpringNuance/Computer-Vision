```python
#preparations
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv

%matplotlib inline
```

```python
img = cv.imread("assets/boat1.png")
print(img.shape)

# convert to a 2d matrix (grayscale)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
print(gray.shape)
```


$\texttt{dst} (x,y) =  \mathrm{det} (M^{(x,y)}) - k  \cdot \left ( \mathrm{tr} (M^{(x,y)} )\right )^2$

```python
fig, axes = plt.subplots(ncols=3, figsize=(16,10))

axes[0].imshow(gray, cmap='gray')
axes[0].set_title("Gray image")

gray = np.float32(gray)
dst = cv.cornerHarris(gray,2,3,0.04)
axes[1].imshow(dst, cmap='gray')
axes[1].set_title("Detect edges")

#result is dilated for marking the corners, not important
dst_dilate = cv.dilate(dst,None)
axes[2].imshow(dst_dilate, cmap='gray')
axes[2].set_title("Dilated (Optional)")

plt.show()
```
